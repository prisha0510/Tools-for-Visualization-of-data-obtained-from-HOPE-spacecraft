```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
#opening the .fits file, modify the path of the file you wish to open
data1 = fits.open(r"C:\Users\prish\Downloads\emm_emu_l2a_20210706t222051_0076_os4a_9734_f_v01-09.fits.gz")
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
sza1=data1['FOV_GEOM'].data['SOLAR_Z_ANGLE'][:,0]
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
ht1=data1['FOV_GEOM'].data['MRH_ALT'][:,0]
test_spa=45
test_int = data1['CAL'].data['CORRECTED_CNTS'].shape[0]//2
```

```python
#function to integrate wavelength data
def integrate_brightness(wavelength, radiance, line_range=[121.56 - 2.0, 121.56+ 2.0]):
    spectral_pos = np.logical_and(line_range[0] < wavelength, wavelength <line_range[1])
    spectral_pos = np.where(spectral_pos)
    wavelength_diff = np.diff(wavelength)
    return np.nansum(radiance[spectral_pos]*wavelength_diff[spectral_pos])
```

```python
integrate_brightness(data1['WAVELENGTH'].data['WAVELENGTH_L2A'][0, test_spa],data1['CAL'].data['RADIANCE'][test_int, test_spa])
rad1 = np.zeros(data1['CAL'].data['RADIANCE'].shape[:2])
for i_int in range(data1['CAL'].data['RADIANCE'].shape[0]):
    for i_spa in range(data1['CAL'].data['RADIANCE'].shape[1]):
        rad1[i_int, i_spa] =integrate_brightness(data1['WAVELENGTH'].data['WAVELENGTH_L2A'][0, i_spa],data1['CAL'].data['RADIANCE'][i_int, i_spa])
```

```python
flat_sza1 = sza1.flatten()
flat_ht1 = ht1.flatten()
flat_rad1=rad1.flatten()
```

```python
#making polar plots with Radius as height, and the theta as solar zenith angle. We plot the radiance data in this plot.
r=flat_ht1
theta=flat_sza1
fig1 = plt.figure()
ax = fig1.add_subplot(projection='polar')
ax.set_rlim(3600,20400)
c = ax.scatter(theta, r,s=5, c=rad1)
```
