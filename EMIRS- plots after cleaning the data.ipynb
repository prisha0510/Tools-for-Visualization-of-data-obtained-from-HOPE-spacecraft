```python
from astropy.io import fits
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
```

```python
#working with level2 data
data3 = fits.open(r"C:\Users\prish\Downloads\emm_emr_l2_20210220t112122_0008_r_v01-04.fits")
data=Table(data3[1].data)
```

```python
lat=data['latitude']
lon=data['longitude'] % 360
bri=data['max_brightness_temp']
alt=data['sc_altitude']
alt_copy=data['sc_altitude']
lt=data['ltime']
```

```python
#taking note of the points where the latitude, longitude and local solar time data are zero, as these are wrong data points
a=np.where(lat==0)
b=np.where(lt==0)
c=np.where(lon==0)
p=a or b or c
```

```python
#removing the wrong data points from the data
bri=np.delete(bri, p)
lat=np.delete(lat,a)
lt=np.delete(lt,b)
lon=np.delete(lon,c)
alt=np.delete(alt,a or b)
```

```python
#plotting the data after cleaning it (removing the wrong data points)
plt.figure(figsize=(14,7))
plt.scatter(lon,lat,c=bri)
plt.ylabel('Latitude (degrees)')
plt.xlabel('Longitude (degrees)')
cbar=plt.colorbar()
cbar.ax.set_ylabel('Max Brightness Temperature (K)')
```
